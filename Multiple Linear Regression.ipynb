# Multiple Linear Regression

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import pandas as pd
import statsmodels.api as sm
```

```python
file = 'Assignment_5_DataSet.xlsx'
```

```python
xl = pd.ExcelFile(file)
```

```python
print(xl.sheet_names)
```

```python
df1 = xl.parse(xl.sheet_names[0])
df2 = xl.parse(xl.sheet_names[1])
df3 = xl.parse(xl.sheet_names[2])
df4 = xl.parse(xl.sheet_names[3])
df5 = xl.parse(xl.sheet_names[4])
df6 = xl.parse(xl.sheet_names[5])
```

## Set 1

```python
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(df1.x,df1.y,alpha=0.5, color='blue')
plt.xlabel('x')
plt.ylabel('y')
title = xl.sheet_names[0] + '_Scatter Plot x vs. y'
plt.title(title)
plt.show()
```

```python
model = ols("y ~ x", data=df1).fit()
print(model.summary())
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "x", fig=fig)
```

The original data meets the requirements for a linear model:
1. the data appears to have a linear relationship.
2. the residuals are nearly normally distributed and centered on 0.

**Model interpretation:**
- R-square: 0.979 the model fits very well the data.
- F-statistic: 4579. It's very large so MSG is way bigger than MSE and, therefore the variance of the data is successfully explained by the model.
- Prob (F-statistic): 4.47e-84. The probability of obtaining the slope/interception produced by the regression assuming the null hypothesis is nearly zero.

- Interception: 0.2381 P-value: 0.61 fail to reject the hypothesis of an interception equal to zero.
- slope: 4.98 p-value = 0 we can reject the null hypothesis, the slope is not 0 and therefore there is a relationship between the variables.

When we force the intercept to be zero, the parameters R-square and slope significance appear to improve, however, this is due the computational of each one.

The model now only can modify its slope, therefore it only has one degree of freedom.

In essence, R-square losses its meaning when we reduce a degree of freedom of the model.

## Set 3

```python
#raw data
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(df3.x,df3.Y,alpha=0.5, color='blue')
plt.xlabel('x')
plt.ylabel('y')
title = xl.sheet_names[2] + '_Scatter Plot x vs. log y'
plt.title(title)
plt.show()
```

```python
model = ols("Y ~ x", data=df3).fit()
print(model.summary())
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "x", fig=fig)
```

The original data does not meet the requirements for a linear model:
1. The data does not appear to have a linear relationship, moreover, it looks like a quadratic relationship.
2. The residuals are not normally distributed and centered on 0, they show a pattern and their variability is not constant along the whole range.

```python
# square root tranformation
df3['sqrt_Y'] = df3['Y'].apply(np.sqrt)
```

```python
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(df3.x,df3.sqrt_Y,alpha=0.5, color='blue')
plt.xlabel('x')
plt.ylabel('y')
title = xl.sheet_names[2] + '_Scatter Plot x vs. sqrt y'
plt.title(title)
plt.show()
```

```python
model = ols("sqrt_Y ~ x", data=df3).fit()
print(model.summary())
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "x", fig=fig)
```

The transformed data meets the requirements for a linear model:

1. the data appear to have a linear relationship.
2. the variability of the residuals is not constant along the whole range, showing smaller variability from 0 to 4, but there are not any noticeable pattern. The residuals are centered on 0.

**Model interpretation:**

The transformation is sqrt y, so

sqrt y = a + b * x

where:
- a = intercept (-2.201)
- b = slope (9.78)

algebraically:

y = a^2 + 2 * a * b + b^2 * x^2

```python
model.params
```

```python
a = model.params[0]
b = model.params[1]
A = a**2 + 2*a*b
print(A)
B = b**2
print(B)
```

Finally, the model is:

y = -38.2 + 95.8 * x^2

The expected value when x = 0 is -38.2 and the relationship is quadratic.

- R-square: 0.857 the regression model accounts for 85.7% of the variance.
- F-statistic: 588. It's large so MSG is bigger than MSE and, therefore the variance of the data is successfully explained by the model.
- Prob (F-statistic): 3.37e-43. The probability of obtaining the slope/interception produced by the regression assuming the null hypothesis is nearly zero.

```python
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(df5.X,df5.Y,alpha=0.5, color='blue')
plt.xlabel('x')
plt.ylabel('y')
title = xl.sheet_names[4] + '_Scatter Plot x vs. y'
plt.title(title)
plt.show()
```

```python
model = ols("Y ~ X", data=df5).fit()
print(model.summary())
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "X", fig=fig)
```

```python
fig, ax = plt.subplots(figsize=(15,3))
fig = sm.graphics.plot_leverage_resid2(model, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
fig = sm.graphics.influence_plot(model, ax=ax)
```

```python
#droping the outlier
df5_ol = df5.drop(100)
```

```python
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(df5_ol.X,df5_ol.Y,alpha=0.5, color='blue')
plt.xlabel('x')
plt.ylabel('y')
title = xl.sheet_names[4] + '_Scatter Plot x vs. y without outliers'
plt.title(title)
plt.show()
```

```python
model = ols("Y ~ X", data=df5_ol).fit()
print(model.summary())
```

The original data meets the requirements for a linear model:

- The data appear to have a linear relationship although there are 3 outliers, one of them remarkably off.
- The residuals are nearly normally distributed and centered on 0.

**Influence of outliers**

The outlier (index = 100) is an influential point with high leverage.

When this point is eliminated the R-square improves from 0.806 to 0.906, as well as the F-statistic.

Since the outlier is on the bottom right corner, it is pulling the line down and therefore the slope with outlier is 4.76 and without it is 5.07.
